```python
product_id = [1,2,3,4,5,6,7,8]
online_price=[236,200,380,345,100,456,345,980]
offline_price=[12,334,56,78,32,655,90,100]
chor_bazar = [100,200,300,400,150,350,250,550]
```

```python
plt.plot(product_id,online_price,color='green', markersize = 10, label="online_price")
plt.plot(product_id,chor_bazar,color='red', marker='+', markersize = 10, label='chor_bazar')
plt.plot(product_id,offline_price,color='yellow', marker='*', markersize = 10, label='offline')

plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product vs Price Comparison")
plt.legend()
plt.show()
```

```python
year=[2001,2002,2003,2004,2005,2006,2007]
price=[20000,30000,67000,12000,34000,23000,9000]
plt.barh(year,price, color ='orange')
plt.xlabel('year')
plt.ylabel('price')
plt.title('comparison of car price by year')
plt.show()
```
